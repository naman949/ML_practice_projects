# Clustering
Clustering is the task of grouping a set of objects in such a way that objects in the same group are more similar to each other than to those in other groups.

## Types of clustering
- ##### Connectivity based clustering
\- Based on the distance between elements. Example: Hierarchical clustering
- ##### Centroid based clustering
\- based on the position of elements. Example: K-Means clustering
- ##### Distribution based clustering
\- Example: Multivariate normal distribution that is used by expectation maximization algorithm
- ##### Density based clustering
\- based on the density. Example: DBSCAN

```python
from sklearn import datasets
```

```python
sns.set_style("darkgrid")
```

```python
sample_data = datasets.make_blobs(n_samples = 200, n_features = 2, cluster_std = 1.2, centers = 3,random_state = 101)
```

```python
sample_data
```

```python
plt.scatter(sample_data[0][:,0],sample_data[0][:,1],c=sample_data[1],cmap='gray')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3,random_state=101)
```

```python
model.fit(sample_data[0])
```

```python
model.labels_
```

```python
plt.scatter(sample_data[0][:,0],sample_data[0][:,1],c=model.labels_,cmap='rainbow')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,sharey=True)
ax1.set_title("K Means Prediction")
ax2.set_title("Original")

ax1.scatter(sample_data[0][:,0],sample_data[0][:,1],c=model.labels_,cmap='rainbow')
ax2.scatter(sample_data[0][:,0],sample_data[0][:,1],c=sample_data[1],cmap='rainbow')
```

```python
from sklearn.cluster import k_means
```
