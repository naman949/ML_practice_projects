```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn import datasets
```

```python
sample_data = datasets.make_blobs(n_samples=300,n_features=3,cluster_std=3,centers=3,random_state=101)
```

```python
sample_data
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

for m in [('o','^')]:
    ax.scatter(sample_data[0][:,0],sample_data[0][:,1],sample_data[0][:,2],c=sample_data[1],marker=('^'),cmap='rainbow')

ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_zlabel('Z label')
```

```python
from sklearn.cluster.hierarchical import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
model.fit(sample_data[0])
```

```python
model.labels_
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

for m in [('o','^')]:
    ax.scatter(sample_data[0][:,0],sample_data[0][:,1],sample_data[0][:,2],c=model.labels_,marker=('^'),cmap='rainbow')

ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_zlabel('Z label')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,sharey=True)
ax1.scatter(sample_data[0][:,0],sample_data[0][:,1],c=model.labels_,marker=('^'),cmap='rainbow')
ax2.scatter(sample_data[0][:,0],sample_data[0][:,1],c=sample_data[1],marker=('^'),cmap='rainbow')
ax1.set_title("Predictions")
ax2.set_title("Original")
```
